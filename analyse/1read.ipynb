```python
# Read the data
df = pd.read_csv('../data/sistema-unico-de-atencion-ciudadana-2021.csv', sep=';')
df.head()
```

```python
#filter for tipo_prestacion
transito = df[df['prestacion'] == 'VEHÍCULO MAL ESTACIONADO']
# transito.to_csv( '../data/transito.csv', index=False)9
transito.head()
```

```python
transito['fecha_ingreso'] = pd.to_datetime(transito['fecha_ingreso'], format='%d/%m/%Y', errors='coerce')
transito['dia_de_la_semana'] = transito['fecha_ingreso'].dt.day_name()
transito.head()
```

```python
df = pd.read_csv('../data/transito_por_hora.csv')
df['hora'] = df.apply(lambda row: row['hora'] %24 , axis=1)
df['numero_dia'] = df.apply(lambda row: "1. Monday" if row['dia'] == "Monday" else "2. Tuesday" if row['dia'] == "Tuesday" else "3. Wednesday" if row['dia'] == "Wednesday" else "4. Thursday" if row['dia'] == "Thursday" else "5. Friday" if row['dia'] == "Friday" else "6. Saturday" if row['dia'] == "Saturday" else "7. Sunday", axis=1)
df.to_csv('fix_transito_por_hora.csv', index=False)
```
